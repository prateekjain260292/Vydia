```python
import numpy as np
import cv2
import time
from sklearn.preprocessing import StandardScaler

v = cv2.VideoCapture("Intern_Video.mp4")
success, image = v.read()


v.set(cv2.CAP_PROP_POS_AVI_RATIO,1)
print(v.get(cv2.CAP_PROP_POS_MSEC))

fps = v.get(cv2.CAP_PROP_FPS)
print(fps)

total = v.get(cv2.CAP_PROP_FPS) * v.get(cv2.CAP_PROP_POS_MSEC) / 1000
print(total)

gap = total / 10
print("sdf")
print(gap)
i = 0
count = 0;
seconds = 5
multiplier = fps * seconds
print(multiplier)
while success:
    frameId = int(round(v.get(1))) #current frame number, rounded b/c sometimes you get frame intervals which aren't integers...this adds a little imprecision but is likely good enough
    success, image = v.read()
    print(frameId)
    if frameId % multiplier == 0:
        cv2.imwrite("image%d.jpg" % i, image)
    i = i+1
    
#################### Setting up the file ################
videoFile = "Intern_Video.mp4"
vidcap = cv2.VideoCapture(videoFile)
success,image = vidcap.read()

#################### Setting up parameters ################

#OpenCV is notorious for not being able to good to 
# predict how many frames are in a video. The point here is just to 
# populate the "desired_frames" list for all the individual frames
# you'd like to capture. 

fps = vidcap.get(cv2.CAP_PROP_FPS)
#est_video_length_minutes = 3         # Round up if not sure.
#est_tot_frames = est_video_length_minutes * 60 * fps  # Sets an upper bound # of frames in video clip

n = 10                             # Desired interval of frames to include
desired_frames = np.arange(0,int(total),int(gap)) 


#################### Initiate Process ################
print(desired_frames)
for i in desired_frames:
    print(i)
    vidcap.set(1,i-1)                      
    success,image = vidcap.read(1)         # image is an array of array of [R,G,B] values
    frameId = vidcap.get(1)                # The 0th frame is often a throw-away
    cv2.imwrite("image%d.jpg" % i, image)

vidcap.release()
print ("Complete")    

```

```python
import numpy as np
import cv2
import time
from sklearn.preprocessing import StandardScaler

cap = cv2.VideoCapture('small.mp4')
#time.sleep(12)

fgbg = cv2.createBackgroundSubtractorMOG2(detectShadows = False)

sol = []

while(1):
    ret, frame = cap.read()

    fgmask = fgbg.apply(frame)
    #print(fgmask)
    #print("sad")
    if fgmask is not None:
        cv2.imshow('frame',fgmask)
    else:
        break
    scaler = StandardScaler()
    scaled = scaler.fit_transform(fgmask)
    sol.append(scaled.mean(axis=0))
    k = cv2.waitKey(30) & 0xff
    if k == 24:
        break
        
solut = scaler.fit_transform(sol)
ans = solut.mean(axis=0)
print(ans)
print(len(ans))

#print()
cap.release()
cv2.destroyAllWindows()
```
